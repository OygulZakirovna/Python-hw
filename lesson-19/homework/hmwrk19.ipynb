```python
# Homework Assignment 1: Analyzing Sales Data

# You are given a dataset containing sales data for an e-commerce website. The dataset (task\sales_data.csv) has the following columns:

# Date: Date of the sale.
# Product: Name of the product sold.
# Category: Category to which the product belongs.
# Quantity: Number of units sold.
# Price: Price per unit.
# Tasks:

# Group the data by the Category column and calculate the following aggregate statistics for each category:
# Total quantity sold.
# Average price per unit.
# Maximum quantity sold in a single transaction.
# Identify the top-selling product in each category based on the total quantity sold.
# Find the date on which the highest total sales (quantity * price) occurred.
```

```python
# 1
import pandas as pd
import numpy as np
df = pd.read_csv('sales_data.csv')
df
```

```python
grouped_by_category = df.groupby('Category')
type(grouped_by_category)
```

```python
grouped_by_category.ngroups
```

```python
for i in grouped_by_category.groups:
    print(i)
```

```python
# Group the data by the Category column and calculate the following aggregate statistics for each category:
# Total quantity sold.
grouped_by_category['Quantity'].agg(sum)
```

```python
# Average price per unit
grouped_by_category['Price'].agg('mean')
```

```python
# Maximum quantity sold in a single transaction.
filtered = grouped_by_category['Quantity'].agg('max')
filtered
```

```python
# Identify the top-selling product in each category based on the total quantity sold.
product_sales = df.groupby(['Category', 'Product'])['Quantity'].sum().reset_index()


sorted_sales = product_sales.sort_values(['Category', 'Quantity'], ascending=[True, False])

top_selling = sorted_sales.groupby('Category').first().reset_index()
top_selling
```

```python
# Find the date on which the highest total sales (quantity * price) occurred.
df['Date'] = pd.to_datetime(df['Date'])

df['TotalSales'] = df['Quantity'] * df['Price']

daily_sales = df.groupby('Date')['TotalSales'].sum()

date_with_highest_sales = daily_sales.idxmax()
date_with_highest_sales 
```

```python
# Homework Assignment 2: Examining Customer Orders

# You have a dataset (task\customer_orders.csv) containing information about customer orders. The dataset has the following columns:

# OrderID: Unique identifier for each order.
# CustomerID: Unique identifier for each customer.
# Product: Name of the product ordered.
# Quantity: Number of units ordered.
# Price: Price per unit.
# Tasks:

# 1 Group the data by CustomerID and filter out customers who have made less than 20 orders.
# 2 Identify customers who have ordered products with an average price per unit greater than $120.
# 3 Find the total quantity and total price for each product ordered, and filter out products that have a total quantity less than 5 units.
```

```python
df = pd.read_csv('customer_orders.csv')
df
```

```python
# # 1 Group the data by CustomerID and filter out customers who have made less than 20 orders.
# filted = df.groupby('CustomerID')
# filted
# for i in filted.groups:
#     print(i)

# certain_customers = filted.count["Quantity"]

# customer_order_counts = df.groupby('CustomerID').size()
# import pandas as pd
# from io import StringIO



```

```python
# 1
customer_order_counts = df.groupby('CustomerID').size().reset_index(name='OrderCount')

frequent_customers = customer_order_counts[customer_order_counts['OrderCount'] <= 20]['CustomerID']


filtered_df = df[df['CustomerID'].isin(frequent_customers)]

print(filtered_df)
```

```python
# 2
average_price_per_customer = df.groupby('CustomerID')['Price'].mean().reset_index(name='AveragePrice')

customers_high_average_price = average_price_per_customer[average_price_per_customer['AveragePrice'] > 120]

high_average_price_customer_ids = customers_high_average_price['CustomerID'].tolist()

filtered_df_high_average_price = df[df['CustomerID'].isin(high_average_price_customer_ids)]

print("Customers with average product price > $120:")
print(customers_high_average_price)

```

```python
# 3 Find the total quantity and total price for each product ordered, and filter out products that have a total quantity less than 5 units.

product_summary = df.groupby('Product').agg(
    TotalQuantity=('Quantity', 'sum'),
    TotalPrice=('Price', 'sum')
).reset_index()

filtered_products = product_summary[product_summary['TotalQuantity'] <= 5]

print("Total quantity and total price per product (with total quantity <= 5):")
print(filtered_products)
```

```python
# Homework Assignment 3: Population Salary Analysis

#1  "task\population.db" sqlite database has population table.
#2 "task\population salary analysis.xlsx" file defines Salary Band categories.
# Read the data from population table and calculate following measures:
# Percentage of population for each salary category;
# Average salary in each salary category;
# Median salary in each salary category;
# Number of population in each salary category;
# 3 Calculate the same measures in each State
# Note: Use SQL only to select data from database. All the other calculations should be done in python.
```

```python
import pandas as pd
import sqlite3

# --- Step 1:
db_path = 'task/population.db'
conn = sqlite3.connect(db_path)
query = "SELECT id, state, city, age, gender, salary FROM population"
population_df = pd.read_sql_query(query, conn)
conn.close()

print("Population Data:")
print(population_df.head())
print("\n")

# --- Step 2: 
excel_path = 'task/population salary analysis.xlsx'
salary_bands_df = pd.read_excel(excel_path)

print("Salary Bands:")
print(salary_bands_df)
print("\n")

# --- Step 3: 
def categorize_salary(salary, bands_df):
    for index, row in bands_df.iterrows():
        if row['Lower Bound'] <= salary < row['Upper Bound']:
            return row['Category']
        elif row['Lower Bound'] == salary and row['Lower Bound'] == bands_df['Lower Bound'].min():
            return row['Category']
    return None  
population_df['Salary Category'] = population_df['salary'].apply(lambda x: categorize_salary(x, salary_bands_df))

print("Population Data with Salary Categories:")
print(population_df.head())
print("\n")

# --- Step 4: 
global_analysis = population_df.groupby('Salary Category').agg(
    PopulationCount=('id', 'count'),
    AverageSalary=('salary', 'mean'),
    MedianSalary=('salary', 'median')
).reset_index()

total_population = population_df['id'].count()
global_analysis['PercentageOfPopulation'] = (global_analysis['PopulationCount'] / total_population) * 100

print("Global Salary Category Analysis:")
print(global_analysis)
print("\n")

# --- Step 5: 
state_analysis = population_df.groupby(['State', 'Salary Category']).agg(
    PopulationCount=('id', 'count'),
    AverageSalary=('salary', 'mean'),
    MedianSalary=('salary', 'median')
).reset_index()

state_total_population = population_df.groupby('State')['id'].count().reset_index(name='TotalStatePopulation')
state_analysis = pd.merge(state_analysis, state_total_population, on='State')
state_analysis['PercentageOfPopulation'] = (state_analysis['PopulationCount'] / state_analysis['TotalStatePopulation']) * 100
state_analysis = state_analysis.drop(columns=['TotalStatePopulation'])

print("Salary Category Analysis by State:")
print(state_analysis)
```
